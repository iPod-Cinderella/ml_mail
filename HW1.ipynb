# Задача 1

Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. 
Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!

Первым выходит прокурор и говорит: 

1.1. “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен - всего 1%" 

1.2. "Значит, с вероятностью 99% он виновен, предлагаю осудить”. 

В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.

## Дано:

Имеем события:

x - виновен

y - группа крови совпадает с найденной на месте убийства.

Из условия знаем:
####  1. Найденная кровь присутствует всего лишь у 1% населения:


* Вероятность встретить группу грови, найденную на месте убийства:

$P(y) = 0,01$

* Вероятность НЕ встретить группу грови, найденную на месте убийства:

$P(\overline{y}) = 0,99$

####  2. Найденная кровь 100% принадлежит убийце, следовательно:


* Вероятность, что группа крови совпадает с найденной на месте убийства при условии, что человек виновен:

$P(y|x) = 1$

* Вероятность, что группа крови НЕ совпадает с найденной на месте убийства при условии, что человек виновен:

$P(\overline{y}|x) = 0$

* Вероятность, что человек виновен, при условии, что его группа крови НЕ совпадает с найденной:

$P(x|\overline{y}) = 0$

* Вероятность, что человек НЕ виновен, при условии, что его группа крови НЕ совпадает с найденной:

$P(\overline{x}|\overline{y}) = 1$

## Решение:

#### 1.1. Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен - всего 1%:

* Прокурор оценил вероятность встретить группу грови, найденную на месте убийства:

$P(y) = 0,01$

* Следуя формулировке утверждения, он должен был оценить вероятность, что группа крови совпадает с найденной на месте убийства при условии, что человек НЕ виновен:

$P(y|\overline{x}) = \frac{P(y) P(\overline{x}|y)}{P(\overline{x})} \approx 0,01 (из п.2)$

####  1.2. "Значит, с вероятностью 99% он виновен, предлагаю осудить”.

* Прокурор оценил веротность НЕ встретить группу грови, найденную на месте убийства:

$P(\overline{y}) = 1 - P(y) = 0,99$

* Следуя формулировке утверждения, он должен был оценить вероятность, что человек виновен, при условии, что его группа крови  совпадает с найденной:

$P(x|y) = \frac{P(x) P(y|x)}{P(y)}$

# Задача 2

Затем выходит адвокат, объясняет ошибки прокурора, а сам говорит следующее: “В городе живёт миллион человек. Это значит, что примерно у 10.000 из них такая группа крови. 

2.1. "Значит, всё, что говорит нам эта кровь - это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. 

В чём не прав адвокат? Укажите, какие вероятности он оценил, а какие должен был оценить.


## Дано:

#### 3. Общее население N = 1млн. 


* Вероятность что человек виновен:

$P(x) = \frac{1}{1.000.000}$

* Вероятность что человек НЕ виновен:

$P(\overline{x}) = 1 - \frac{1}{1.000.000}$

## Решение:

Адвокат оценил вероятность, что человек виновен, при условии, что его группа крови  совпадает с найденной:

$P(x|y) = \frac{P(x) P(y|x)}{P(y)}$

НО исключать группу крови из материалов дела нельзя, т.к. в совокупности с другими уликами, совпадение группы крови, может значительно увеличить вероятность виновности подсудимого
