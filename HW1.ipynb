# Задача 1

Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. 
Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!

Первым выходит прокурор и говорит: 

1.1. “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен - всего 1%" 

1.2. "Значит, с вероятностью 99% он виновен, предлагаю осудить”. 

В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.

## Дано:

Имеем события:

x - виновен

y - группа крови совпадает с найденной на месте убийства.

Из условия знаем:
####  1. Найденная кровь присутствует всего лишь у 1% населения:


* Вероятность встретить группу грови, найденную на месте убийства:

$P(y) = 0,01$

* Вероятность НЕ встретить группу грови, найденную на месте убийства:

$P(\overline{y}) = 0,99$

####  2. Найденная кровь 100% принадлежит убийце, следовательно:


* Вероятность, что группа крови совпадает с найденной на месте убийства при условии, что человек виновен:

$P(y|x) = 1$

* Вероятность, что группа крови НЕ совпадает с найденной на месте убийства при условии, что человек виновен:

$P(\overline{y}|x) = 0$

* Вероятность, что человек виновен, при условии, что его группа крови НЕ совпадает с найденной:

$P(x|\overline{y}) = 0$

* Вероятность, что человек НЕ виновен, при условии, что его группа крови НЕ совпадает с найденной:

$P(\overline{x}|\overline{y}) = 1$

## Решение:

#### 1.1. Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен - всего 1%:

* Прокурор оценил вероятность встретить группу грови, найденную на месте убийства:

$P(y) = 0,01$

* Следуя формулировке утверждения, он должен был оценить вероятность, что группа крови совпадает с найденной на месте убийства при условии, что человек НЕ виновен:

$P(y|\overline{x}) = \frac{P(y) P(\overline{x}|y)}{P(\overline{x})} \approx 0,01 (из п.2)$

####  1.2. "Значит, с вероятностью 99% он виновен, предлагаю осудить”.

* Прокурор оценил веротность НЕ встретить группу грови, найденную на месте убийства:

$P(\overline{y}) = 1 - P(y) = 0,99$

* Следуя формулировке утверждения, он должен был оценить вероятность, что человек виновен, при условии, что его группа крови  совпадает с найденной:

$P(x|y) = \frac{P(x) P(y|x)}{P(y)}$

# Задача 2

Затем выходит адвокат, объясняет ошибки прокурора, а сам говорит следующее: “В городе живёт миллион человек. Это значит, что примерно у 10.000 из них такая группа крови. 

2.1. "Значит, всё, что говорит нам эта кровь - это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. 

В чём не прав адвокат? Укажите, какие вероятности он оценил, а какие должен был оценить.


## Дано:

#### 3. Общее население N = 1млн. 


* Вероятность что человек виновен:

$P(x) = \frac{1}{1.000.000}$

* Вероятность что человек НЕ виновен:

$P(\overline{x}) = 1 - \frac{1}{1.000.000}$

## Решение:

Адвокат оценил вероятность, что человек виновен, при условии, что его группа крови  совпадает с найденной:

$P(x|y) = \frac{P(x) P(y|x)}{P(y)}$

НО исключать группу крови из материалов дела нельзя, т.к. в совокупности с другими уликами, совпадение группы крови, может значительно увеличить вероятность виновности подсудимого

# Задача 3

Коронавирусом по нашему предположению болеет 1% населения, а тест имеет вероятность ошибки 5% в обе стороны. 

Мы видели, что в результате человек с положительным тестом имеет апостериорную вероятность оказаться больным всего лишь около 16%.

Врачи решили, что терпеть это никак нельзя; они готовы провести срочное исследование и улучшить качество теста на коронавирус. Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт позитивный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного).

Помогите медикам: постройте зависимости апостериорных вероятностей болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?

Имеем события:

x - человек болен

y - тест показал положительный результат

$P(x) = 0,01$

Вероятность, что человек болен при положительном тесте:
$P(x|y) = \frac{P(x) P(y|x)}{P(y)} = \frac{P(x) P(y|x)}{P(x) P(y|x)  +  P(\overline{x}) P(y|\overline{x})}$

Вероятность, что человек болен при отрицательном тесте:
$P(x|\overline{y}) = \frac{P(x) P(\overline{y}|x)}{P(\overline{y})} = \frac{P(x) P(\overline{y}|x)}{P(x) P(\overline{y}|x)  +  P(\overline{x}) P(\overline{y}|\overline{x})}$

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import json
import numpy as np
import scipy as sp
import scipy.stats as st
import scipy.integrate as integrate
from sklearn import linear_model

sns.set_style("whitegrid")
sns.set_palette("colorblind")
palette = sns.color_palette()
figsize = (15,8)
legend_fontsize = 16
```

```python
## если data = 1, тест положительный
def compute_posterior(data, false_positives = 0.05, false_negatives = 0.05, prior_prob = 0.01):
    if data == 1:
        return (prior_prob * (1 - false_negatives)) / (prior_prob * (1 - false_negatives) + (1 - prior_prob) * false_positives)
    elif data == 0:
        return (prior_prob * false_negatives) / (prior_prob * false_negatives + (1 - prior_prob) * (1 - false_positives))
    else:
        return None

vector_posterior = np.vectorize(compute_posterior, excluded=['data', 'test_success'])
xs = np.arange(0, 0.05, 0.001)
ys_f_p_one = vector_posterior(1, false_positives = 0.05 - xs)
ys_f_n_one = vector_posterior(1, false_negatives = 0.05 - xs)

ys_f_p_zero = vector_posterior(0, false_positives = 0.05 - xs)
ys_f_n_zero = vector_posterior(0, false_negatives = 0.05 - xs)
```

```python
print("Видим, что при положительном тесте, апостериорная вероятность значительно зависит от изменения false_positives, в то время, как изменение false_negatives дает ничтожное улучшение.")


print("\nАпостериорная вероятность после положительного теста:")
print("false_positives = 0.05, false_negatives = 0.05: %.6f" % compute_posterior(1))

print("\nfalse_positives = 0.01, false_negatives = 0.05: %.6f" % compute_posterior(1, false_positives = 0.01))
print("Увеличение в: %.3f" % (compute_posterior(1, false_positives = 0.01)/compute_posterior(1)))

print("\nfalse_positives = 0.05, false_negatives = 0.01: %.6f" % compute_posterior(1, false_negatives = 0.01))
print("Увеличение в: %.3f" % (compute_posterior(1, false_negatives = 0.01)/compute_posterior(1)))

fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(111)
ax.plot(xs, ys_f_p_one, linewidth=2, label='Изменение false_positives')
ax.plot(xs, ys_f_n_one, linewidth=2, label='Изменение false_negatives')
ax.set_xlim((0.0, 0.05))
ax.set_ylabel('Апостериорная вероятностьпосле положительного теста', fontsize=legend_fontsize)
ax.set_xlabel('delta', fontsize=legend_fontsize)
ax.legend(fontsize=legend_fontsize)
plt.show()
```

```python
print("Видим, что при отрицательном тесте, апостериорная вероятность сильнее зависит от изменения false_negatives, но так же присутсвует улучшение и при изменении false_positives")


print("\nАпостериорная вероятность после отрицательного теста:")
print("false_positives = 0.05, false_negatives = 0.05: %.6f" % compute_posterior(0))

print("\nfalse_positives = 0.01, false_negatives = 0.05: %.6f" % compute_posterior(0, false_positives = 0.01))
print("Снижение в: %.3f" % (compute_posterior(0, false_positives = 0.01)/compute_posterior(0)))

print("\nfalse_positives = 0.05, false_negatives = 0.01: %.6f" % compute_posterior(0, false_negatives = 0.01))
print("Снижение в: %.3f" % (compute_posterior(0, false_negatives = 0.01)/compute_posterior(0)))

fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(111)
ax.plot(xs, ys_f_p_zero, linewidth=2, label='Изменение false_positives')
ax.plot(xs, ys_f_n_zero, linewidth=2, label='Изменение false_negatives')
ax.set_xlim((0.0, 0.05))
ax.set_ylabel('Апостериорная вероятность после отрицательного теста', fontsize=legend_fontsize)
ax.set_xlabel('delta', fontsize=legend_fontsize)
ax.legend(fontsize=legend_fontsize)
plt.show()
```

 С учетом построенных графиков, измение false_positives дает больший вклад в улучшение показателей. Поэтому медикам стоит сконцентрироваться на этом показателе
